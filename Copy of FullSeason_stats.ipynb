```python
# Import scraping modules
from urllib.request import urlopen
from bs4 import BeautifulSoup

# Import data manipulation modules
import pandas as pd
import numpy as np

# Import data visualization modules
import matplotlib as mpl
import matplotlib.pyplot as plt

from google.colab import files


pd.set_option('display.max_columns', None)

url = f"https://www.baseball-reference.com/awards/hof_batting.shtml"

df = pd.read_html(url)[0]
df["Inducted"] = pd.to_datetime(df["Inducted"], format='%Y')
```

```python
download = df[df["Inducted"]>='1990-01-01']
download.to_csv('mlb_hof_stats.csv')
files.download('mlb_hof_stats.csv')
```

```python
df_list = []
years = range(1966, 1980)
for year in years:
  # URL of page
  print(year)
  url = f"https://www.pro-football-reference.com/years/{year}/passing.htm"
  df = pd.read_html(url)[0]
  df['Years'] = year
  df_list.append(df)

results = pd.concat(df_list)
results = results[results['Rk'].ne('Rk')]
```

```python
QB_results = results[results["Pos"]=='QB'].sort_values(["Player","Years"])
QB_results["Player"] = QB_results["Player"].replace('*', '')
QB_results["Player"] = QB_results["Player"].replace('+', '')
```

```python
record= QB_results['QBrec'].str.split('-',n=2,expand = True)
record[0]
```

```python
QB_results['Wins']=record[0]
QB_results['Losses']=record[1]
QB_results['Ties']=record[2]
```

```python
QB_results = QB_results.drop(columns=['QBrec'])
QB_results = QB_results[['Rk', 'Player', 'Tm', 'Age', 'Pos', 'G', 'GS', 'Wins', 'Losses', 'Ties', 'Cmp', 'Att', 'Cmp%', 'Yds', 'TD', 'TD%', 'Int', 'Int%', '1D', 'Lng', 'Y/A', 'AY/A', 'Y/C', 'Y/G', 'Rate', 'Sk', 'Yds.1', 'Sk%', 'NY/A', 'ANY/A', '4QC', 'GWD', 'Years', 'QBR']]
```

```python
df2_list = []
years = range(2000, 2023)
for year in years:
  # URL of page
  print(year)
  url2 = f"https://www.pro-football-reference.com/years/{year}/rushing.htm"
  df2 = pd.read_html(url2,header=1)[0]
  df2['Years'] = year
  df2_list.append(df2)

results2 = pd.concat(df2_list)
results2 = results2[results2['Rk'].ne('Rk')]
```

```python
RB_results = results2[results2["Pos"]=='RB'].sort_values(["Player","Years"])
RB_results["Player"] = RB_results["Player"].replace('*', '')
RB_results["Player"] = RB_results["Player"].replace('+', '')
```

```python
RB_results.head()
```

```python
RB_results.to_csv('rb_stats.csv')
files.download('rb_stats.csv')
```
